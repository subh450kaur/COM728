```python
import pandas as pd
from data_loading import load_data, load_data_pandas
from data_analysis import get_top_job_roles_by_education, calculate_average_monthly_rate, calculate_average_employment_duration, get_top_10_satisfied_employees
from data_retrive import retrieve_employee_info, retrieve_job_role_info, retrieve_employees_with_low_standard_hours, retrieve_employees_with_high_satisfaction
from data_visualisation import  proportion_of_employees_by_education,employee_training_frequency,job_satisfaction_across_levels,average_monthly_income_by_role

# File path and data initialization
file_location = 'HR_Metrics.csv'
data = load_data(file_location)
HR_data = load_data_pandas(file_location)

def print_table(data):
    if isinstance(data, pd.DataFrame):
        if data.empty:
            print("No data to display.")
        else:
            print(data)
    elif isinstance(data, list):
        if not data:
            print("No data to display.")
        else:
            df = pd.DataFrame(data)
            print(df)
    else:
        print("Invalid data format. Cannot display the table.")

# Task A - Data Retrieval
def retrieve_data_task_a():
    while True:
        print("\nTask a) Retrieve Data:")
        print("a1. Retrieve Gender, Age, Marital Status, and Department by Employee Number")
        print("a2. Retrieve Department, Education Level, Total Working Years, and Standard Hours by Job Role")
        print("a3. Retrieve Employee Number, Job Role, Marital Status, and Hourly Rate by Department (Standard Hours < 60)")
        print("a4. Retrieve Employee Number, Monthly Income, and Years at Company by Job Role (High Satisfaction)")
        print("0. Return to main menu")
        choice = input("Enter your choice: ")

        if choice == 'a1':
            emp_number = input("Enter the Employee Number: ")
            result = retrieve_employee_info(data, emp_number)
            if result:
                print("\nGender: ", result[0])
                print("Age: ", result[1])
                print("Marital Status: ", result[2])
                print("Department: ", result[3])
            else:
                print("No employee found with the given Employee Number.")
        
        elif choice == 'a2':
            job_role = input("Enter the Job Role: ")
            results = retrieve_job_role_info(data, job_role)
            if results:
                print("\nDepartment | Education Level | Total Working Years | Standard Hours")
                for row in results:
                    print(" | ".join(map(str, row)))
            else:
                print("No data found for the specified Job Role.")
        
        elif choice == 'a3':
            department_name = input("Enter the Department Name: ")
            results = retrieve_employees_with_low_standard_hours(data, department_name)
            if results:
                print("\nEmployee Number | Job Role | Marital Status | Hourly Rate")
                for row in results:
                    print(" | ".join(map(str, row)))
            else:
                print("No employees found in the specified Department with Standard Hours < 60.")
        
        elif choice == 'a4':
            job_role = input("Enter the Job Role: ")
            results = retrieve_employees_with_high_satisfaction(data, job_role)
            if results:
                print("\nEmployee Number | Monthly Income | Years at Company")
                for row in results:
                    print(" | ".join(map(str, row)))
            else:
                print("No employees found with high satisfaction levels for the specified Job Role.")
        
        elif choice == '0':
            break
        else:
            print("Invalid choice. Please try again.")


# Task B - Data Analysis
def analyze_data_task_b(HR_data):
    while True:
        print("\nTask B) Analyze Data:")
        print("b1. Identify the top 3 job roles with the highest number of employees hired based on education level")
        print("b2. Analyze the average monthly rate of employees for a specific education level")
        print("b3. Analyze the average duration of employment for employees in a specific department")
        print("b4. Analyze top 10 satisfied employees not promoted in the last 5 years")
        print("0. Return to main menu")
        choice = input("Enter your choice: ")

        if choice == 'b1':
            results = get_top_job_roles_by_education(HR_data)
            print("\nTop 3 Job Roles by Education Level:")
            print(results)

        elif choice == 'b2':
            average_rate = calculate_average_monthly_rate(HR_data)
            print(average_rate)

        elif choice == 'b3':
            average_duration = calculate_average_employment_duration(HR_data)
            print(f"Average Duration : {average_duration}")

        elif choice == 'b4':
            top_employees = get_top_10_satisfied_employees(HR_data)
            print("\nTop 10 Satisfied Employees Not Promoted in Last 5 Years:")
            print(top_employees)

        elif choice == '0':
            break
        else:
            print("Invalid choice. Please try again.")


def visualize_data_task_c():
    """
    Menu-driven function for Task C: Visualize Data.
    Users can choose from four visualization options or return to the main menu.
    """
    while True:
        print("\n--- Task C: Data Visualization ---")
        print("c1. Show the proportion of employees based on their education levels")
        print("c2. Compare the frequency of employee training across different education fields")
        print("c3. Analyze job satisfaction scores across job levels for each department")
        print("c4. Visualize the average monthly income by job role")
        print("0. Return to the main menu")
        
        choice = input("\nEnter your choice (0-4): ")

        # Execute the corresponding visualization function based on user choice
        if choice == 'c1':
            print("\nVisualizing the proportion of employees by education level...")
            proportion_of_employees_by_education(HR_data)
        elif choice == 'c2':
            print("\nComparing the frequency of employee training by education field...")
            employee_training_frequency(HR_data)
        elif choice == 'c3':
            print("\nAnalyzing job satisfaction scores across job levels by department...")
            job_satisfaction_across_levels(HR_data)
        elif choice == 'c4':
            print("\nVisualizing the average monthly income by job role...")
            average_monthly_income_by_role(HR_data)
        elif choice == '0':
            print("\nReturning to the main menu...")
            break
        else:
            print("\nInvalid choice. Please enter a valid option (0-4).")

# Main interaction loop
def main():
    while True:
        print("\n--- Main Menu ---")
        print("a. Retrieve Employee and Job Information (Task A)")
        print("b. Analyze Employee Data Trends (Task B)")
        print("c. Visualize Employee Metrics (Task C)")
        print("0. Exit")
        choice = input("Enter your choice: ")


        if choice == 'a':
            retrieve_data_task_a()
        elif choice == 'b':
            analyze_data_task_b(HR_data)
        elif choice == 'c':
            visualize_data_task_c()
        elif choice == '0':
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
